# **Module 4: Data Summarization and Visualization**

# Overview

This module covers statistical summary measures, data elaboration, and 1-D, 2-D, and n-D statistical data analysis, as well as various visualization techniques including scatter plots, dot charts, and bar plots.

# 1. Statistical Summary Measures

```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load the dataset
url = 'https://raw.githubusercontent.com/salemprakash/EDA/main/Data/HepatitisCdata.csv'
data = pd.read_csv(url)

# Displaying the statistical summary of the dataset
summary = data.describe()
summary
```

# 2. Data Elaboration

```python
# Select only numeric columns for analysis
numeric_data = data.select_dtypes(include=[np.number])

# Calculate additional statistical measures
data_summary = pd.DataFrame({
    'Mean': numeric_data.mean(),
    'Median': numeric_data.median(),
    'Variance': numeric_data.var(),
    'Standard Deviation': numeric_data.std(),
    'Skewness': numeric_data.skew(),
    'Kurtosis': numeric_data.kurt()
})

data_summary
```

# 3. 1-D Statistical Data Analysis

```python
# 1-D Analysis: Distribution plots for 'Age'
plt.figure(figsize=(10, 6))
sns.histplot(data['Age'], bins=20, kde=True, color='blue')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Create a skewness plot for 'ALT' (Alanine Aminotransferase)
plt.figure(figsize=(10, 6))
sns.histplot(data['ALT'], kde=True, bins=30, color='blue')
plt.title("Skewness Plot of ALT Levels")
plt.xlabel("ALT Levels")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```

# 4. 2-D Statistical Data Analysis

```python
# 2-D Analysis: Scatter plot for 'Age' vs 'ALB' (Albumin level)
plt.figure(figsize=(8, 5))
plt.scatter(data['Age'], data['ALB'], color='red')
plt.title('Scatter Plot: Age vs Albumin Level')
plt.xlabel('Age')
plt.ylabel('Albumin Level (ALB)')
plt.grid(True)
plt.show()
```

# 5. n-D Statistical Data Analysis

```python
# Selecting relevant numerical columns for pair plot
selected_columns = ['Age', 'ALB', 'ALT', 'AST']

# Creating pair plot
sns.pairplot(data[selected_columns], diag_kind='kde')
plt.suptitle('Pair Plot of Multiple Variables', y=1.02)
plt.show()

```

# 6. Contingency Tables

```python
# Creating a contingency table for 'Sex' and 'Category'
contingency_table = pd.crosstab(data['Sex'], data['Category'])
print(contingency_table)
```

# 7. Visualization: Scatter Plots, Dot Charts, and Bar Plots

```python
# Scatter plot for 'Age' vs 'BIL' (Bilirubin level)
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Age', y='BIL', data=data, color='green')
plt.title('Scatter Plot: Age vs Bilirubin Level')
plt.xlabel('Age')
plt.ylabel('Bilirubin Level (BIL)')
plt.grid(True)
plt.show()
```

```python
# Dot chart for 'Category' vs 'ALT' (Alanine Aminotransferase level)
plt.figure(figsize=(8, 5))
plt.plot(data['Category'], data['ALT'], 'o', color='purple')
plt.title('Dot Chart: Disease Category vs ALT Levels')
plt.xlabel('Disease Category')
plt.ylabel('ALT Level')
plt.grid(True)
plt.show()
```

```python
# Bar plot for 'Category' vs 'ALT' (Alanine Aminotransferase level)
plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='ALT', data=data, palette='viridis')
plt.title('Bar Plot: Disease Category vs ALT Levels')
plt.xlabel('Disease Category')
plt.ylabel('ALT Level')
plt.grid(True)
plt.show()
```

```python
import ternary
import matplotlib.pyplot as plt

# Selecting relevant numerical columns
data_subset = data[['ALT', 'AST', 'BIL']].dropna()

# Normalize Data: Convert values to proportions summing to 1
data_values = data_subset.values
row_sums = data_values.sum(axis=1, keepdims=True)
normalized_data = data_values / row_sums  # Normalize each row

# Create a ternary plot
scale = 1  # Data is now proportion-based
figure, tax = ternary.figure(scale=scale)
tax.boundary(linewidth=2.0)  # Draw boundary
tax.gridlines(color="gray", multiple=0.1)  # Gridlines for readability

# Plot normalized points
tax.scatter(list(map(tuple, normalized_data)), color='red', s=50, label="Patients")

# Set Labels
tax.left_corner_label("ALT", fontsize=12)
tax.right_corner_label("AST", fontsize=12)
tax.top_corner_label("BIL", fontsize=12)

# Ticks and Styling
tax.ticks(axis='lbr', multiple=0.1, linewidth=1, fontsize=10)
tax.legend()
tax.show()
```
